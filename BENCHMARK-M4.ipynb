```python
# Per ogni continente trovare il numero di casi totali avvenuti in quello stesso continente

# Filtro il dataset per le righe con valori non nulli nella colonna "continent"
continent_cases = strutture[strutture['continent'].notnull()]

# Raggruppo e sommo i nuovi casi per ogni continente (delta giornaliero)
total_new_cases_per_continent = strutture.groupby('continent')['new_cases'].sum()

print("Nuovi casi totali per continente:")
print(total_new_cases_per_continent)
```

```python
# Eseguo la consegna calcolando anche le locazioni senza indicazione di continente

def confronta_casi_totali(dataset, continente1, continente2):
    
    # Estraggo i casi totali per i due continenti specificati
    casi_continente1 = dataset[dataset['continent'] == continente1]['new_cases']
    casi_continente2 = dataset[dataset['continent'] == continente2]['new_cases']
    
    # Calcolo la somma totale dei casi
    casi_totali = dataset['new_cases'].sum()
    
    # Calcolo i descrittori statistici per il continente 1
    min_casi_continente1 = casi_continente1.min()
    max_casi_continente1 = casi_continente1.max()
    mean_casi_continente1 = casi_continente1.mean()
    perc_casi_continente1 = (casi_continente1.sum() / casi_totali) * 100
    
    # Calcolo i descrittori statistici per il continente 2
    min_casi_continente2 = casi_continente2.min()
    max_casi_continente2 = casi_continente2.max()
    mean_casi_continente2 = casi_continente2.mean()
    perc_casi_continente2 = (casi_continente2.sum() / casi_totali) * 100
    
    # Creo un dizionario per i risultati
    risultati = {
        'Continente 1': continente1,
        'Minimo Casi Continente 1': min_casi_continente1,
        'Massimo Casi Continente 1': max_casi_continente1,
        'Media Casi Continente 1': mean_casi_continente1,
        'Percentuale Casi Continente 1': perc_casi_continente1,
        'Continente 2': continente2,
        'Minimo Casi Continente 2': min_casi_continente2,
        'Massimo Casi Continente 2': max_casi_continente2,
        'Media Casi Continente 2': mean_casi_continente2,
        'Percentuale Casi Continente 2': perc_casi_continente2
    }
    
    return risultati

# Chiedo all'utente i nomi dei continenti di interesse
continente1 = input("Inserisci il nome del primo continente,a scelta tra i seguenti Africa, Asia, Europe, North America, Oceania, South America: ")
continente2 = input("Inserisci il nome del secondo continente,a scelta tra i seguenti Africa, Asia, Europe, North America, Oceania, South America: ")

# Utilizzo della funzione
risultati = confronta_casi_totali(strutture, continente1, continente2)
print(risultati)
```

```python
# Eseguo la consegna calcolando anche le locazioni senza indicazione di continente

def confronta_vaccini_totali(dataset, continente1, continente2):
    
    # Estraggo i vaccini totali per i due continenti specificati
    vaccini_continente1 = dataset[dataset['continent'] == continente1]['new_vaccinations']
    vaccini_continente2 = dataset[dataset['continent'] == continente2]['new_vaccinations']
    
    # Calcolo la somma totale dei casi
    vaccini_totali = dataset['new_vaccinations'].sum()
    
    # Calcolo i descrittori statistici per il continente 1
    min_vaccini_continente1 = vaccini_continente1.min()
    max_vaccini_continente1 = vaccini_continente1.max()
    mean_vaccini_continente1 = vaccini_continente1.mean()
    perc_vaccini_continente1 = (vaccini_continente1.sum() / vaccini_totali) * 100
    
    # Calcolo i descrittori statistici per il continente 2
    min_vaccini_continente2 = vaccini_continente2.min()
    max_vaccini_continente2 = vaccini_continente2.max()
    mean_vaccini_continente2 = vaccini_continente2.mean()
    perc_vaccini_continente2 = (vaccini_continente2.sum() / vaccini_totali) * 100
    
    # Converto i valori in formato leggibile utilizzando humanize
    min_vaccini_continente1 = humanize.intcomma(min_vaccini_continente1)
    max_vaccini_continente1 = humanize.intcomma(max_vaccini_continente1)
    mean_vaccini_continente1 = humanize.intcomma(int(mean_vaccini_continente1))
    perc_vaccini_continente1 = humanize.intword(perc_vaccini_continente1)
    min_vaccini_continente2 = humanize.intcomma(min_vaccini_continente2)
    max_vaccini_continente2 = humanize.intcomma(max_vaccini_continente2)
    mean_vaccini_continente2 = humanize.intcomma(int(mean_vaccini_continente2))
    perc_vaccini_continente2 = humanize.intword(perc_vaccini_continente2)
    
    # Creo un dizionario per i risultati
    risultati = {
        'Continente 1': continente1,
        'Minimo Vaccini eseguiti in una giornata Continente 1': min_vaccini_continente1,
        'Massimo Vaccini eseguiti in una giornata Continente 1': max_vaccini_continente1,
        'Media Vaccini eseguiti in una giornata Continente 1': mean_vaccini_continente1,
        'Percentuale Vaccini in Continente 1': perc_vaccini_continente1,
        'Continente 2': continente2,
        'Minimo Vaccini eseguiti in una giornata Continente 2': min_vaccini_continente2,
        'Massimo Vaccini eseguiti in una giornata Continente 2': max_vaccini_continente2,
        'Media Vaccini eseguiti in una giornata Continente 2': mean_vaccini_continente2,
        'Percentuale Vaccini in Continente 2': perc_vaccini_continente2
    }
    
    return risultati

# Chiedo all'utente i nomi dei continenti di interesse
continente1 = input("Inserisci il nome del primo continente,a scelta tra i seguenti Africa, Asia, Europe, North America, Oceania, South America: ")
continente2 = input("Inserisci il nome del secondo continente,a scelta tra i seguenti Africa, Asia, Europe, North America, Oceania, South America: ")

# Utilizzo della funzione
risultati = confronta_vaccini_totali(strutture, continente1, continente2)
print(risultati)
```

```python
import matplotlib.pyplot as plt

# Funzione per disegnare un grafico a torta
def disegna_grafico_a_torta(dati, titolo):
    etichette = dati.keys()  # Etichette
    dimensioni = dati.values()  # Dimensioni
    esplosione = (0.1, 0, 0)  # Esplosione della fetta più grande

    plt.figure(figsize=(8, 6))  # Dimensioni del grafico
    plt.pie(dimensioni, explode=esplosione, labels=etichette, autopct='%1.1f%%', startangle=140)  # Disegno del grafico a torta
    plt.axis('equal') 
    plt.title(titolo)  # Titolo del grafico
    plt.show()  # Mostra il grafico

# Calcolo delle percentuali di casi totali e vaccini totali per i tre continenti selezionati
percentuali_casi = {continente: descr['Percentuale Casi'] for continente, descr in risultati_casi.items()}
percentuali_vaccini = {continente: descr['Percentuale Vaccini'] for continente, descr in risultati_vaccini.items()}

# Disegno dei grafici a torta
disegna_grafico_a_torta(percentuali_casi, 'Percentuale di casi totali per continente')
disegna_grafico_a_torta(percentuali_vaccini, 'Percentuale di vaccini totali per continente')
```
